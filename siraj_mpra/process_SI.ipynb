```python
"""
This notebook contains code to download and process MPRA data from Siraj et al. (2024)
https://www.biorxiv.org/content/10.1101/2024.05.05.592437v1"

The processed data are provided as part of this github repo.
"""
```

```python
import pandas as pd
```

```python
print("Run in console or uncomment os.system call")
print("wget https://www.biorxiv.org/content/biorxiv/early/2024/05/06/2024.05.05.592437/DC4/embed/media-4.xlsx")
# import os; os.system("wget https://www.biorxiv.org/content/biorxiv/early/2024/05/06/2024.05.05.592437/DC4/embed/media-4.xlsx")
```

```python
df = pd.read_excel("media-4.xlsx", engine="openpyxl", skiprows=1)
```

```python
col = ["variant", "active.any", "emVar.any"] + [c for c in df.columns if c.endswith("K562")]
k562 = df[col]
k562.to_csv("media-4-K562_allelic_mpra.tsv.gz", sep="\t", index=False)
```
