```python
import pandas as pd
import numpy as np
```

```python
print("Run in console or uncomment os.system call")
print("wget https://www.biorxiv.org/content/biorxiv/early/2024/05/06/2024.05.05.592437/DC4/embed/media-4.xlsx")
# import os; os.system("wget https://www.biorxiv.org/content/biorxiv/early/2024/05/06/2024.05.05.592437/DC4/embed/media-4.xlsx")
```

```python
df = pd.read_excel("media-4.xlsx", engine="openpyxl", skiprows=1)
```

```python
col = ["variant", "active.any", "emVar.any"] + [c for c in df.columns if c.endswith("K562")]
k562 = df[col]
k562.to_csv("media-4-K562_allelic_mpra.tsv.gz", sep="\t", index=False)
```

```python
oligos = pd.read_excel("media-3.xlsx", engine="openpyxl", skiprows=1)
oligos[["chrom", "pos", "a1", "a2"]] = oligos["Variant"].str.split(':', expand=True)
oligos.to_csv("media-3_oligos.tsv.gz", sep="\t", index=False)
```

```python
oligos_snps = oligos[(oligos["a1"].str.len() == 1) * (oligos["a2"].str.len() == 1)]
oligos_snps.to_csv("media-3_oligos_snps.tsv.gz", sep="\t", index=False)
```

```python
norm = np.log2((df["mean_RNA_ref_K562"] / df["mean_Plasmid_ref_K562"]) / (df["mean_RNA_alt_K562"] / df["mean_Plasmid_alt_K562"]))
unnorm = np.log2((df["mean_RNA_ref_K562"]) / (df["mean_RNA_alt_K562"]))
```

```python
np.abs((norm).loc[df["emVar_K562"] == 1]).min()
```

```python
sns.distplot((norm).loc[df["emVar_K562"] == 1])
sns.distplot((norm).loc[df["emVar_K562"] == 0])
```

```python
sns.kdeplot((unnorm).loc[df["emVar_K562"] == 1])
sns.kdeplot((unnorm).loc[df["emVar_K562"] == 0])
```
